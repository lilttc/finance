 ### Little scripts visualising payoff curves of some basic derivatives and basic strategies.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Forward

```python
K = 100 # Strike Price
S = np.linspace(90,110,100)
long_fwd=S-K
short_fwd=K-S

plt.plot(S, long_fwd, label="long forward",ls="-", marker="None")
plt.plot(S, short_fwd, label="short forward",ls="-", marker="None")
plt.xlabel("Underlying asset price at maturity - $S_T$")
plt.ylabel("Payoff at maturity - $V_T$")
plt.xticks(np.arange(90, 115, 5))
plt.yticks(np.arange(-10, 15, 5))
plt.legend()
plt.title('Payoff Curve of Forward')
plt.axhline(y=0,ls="-",color='black')
```

 ### 2. Call options

```python
K = 100 # Strike Price
S = np.linspace(90,110,100)
Vcall = np.maximum((S-K),0.)

plt.plot(S, Vcall, label="Long Call",ls="-", marker="None")
plt.plot(S, -Vcall, label="Short Call",ls="-", marker="None")
plt.xlabel("Underlying asset price at maturity - $S_T$")
plt.ylabel("Payoff at maturity - $V_T$")
plt.xticks(np.arange(90, 115, 5))
plt.yticks(np.arange(-10, 15, 5))
plt.title('Payoff Curve of Call Option')
plt.legend()
plt.axhline(y=0,ls="-",color='black')
```

 ### 3. Put Options

```python
K = 100 # Strike Price
S = np.linspace(90,110,100)

Vput = np.maximum((K-S),0.)
Vstraddle = Vcall+Vput
plt.plot(S, Vput, label="long put",ls="-", marker="None")
plt.plot(S,-Vput, label="short put",ls="-", marker="None")
plt.xlabel("Underlying asset price at maturity - $S_T$")
plt.ylabel("Payoff at maturity - $V_T$")
plt.xticks(np.arange(90, 115, 5))
plt.yticks(np.arange(-10, 15, 5))
plt.title("Payoff curve of Put Option")
plt.legend()
plt.axhline(y=0,ls="-",color='black')
```

 ### 4. Straddle Strategy

```python
K = 100 # Strike Price
S = np.linspace(90,110,100)
Vcall = np.maximum((S-K),0.)
Vput = np.maximum((K-S),0.)
Vstraddle = Vcall+Vput
plt.plot(S, Vstraddle, label="Long straddle")
plt.xlabel("Underlying asset price at maturity - $S_T$")
plt.ylabel("Payoff at maturity - $V_T$")
plt.xticks(np.arange(90, 115, 5))
plt.yticks(np.arange(0, 15, 5))
plt.title('Payoff Curve of Straddle Strategy')
```
